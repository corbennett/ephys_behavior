```python
mouse_num = '422699'


json_string = request.urlopen("http://lims2/specimens/isi_experiment_details/" + mouse_num + ".json").read()
info = json.loads(json_string)

experiments = info[0]['isi_experiments']

#get only experiments that passed QC
passed = [i for i, e in enumerate(experiments) if e['workflow_state'] == 'passed']
experiments = [experiments[p] for p in passed]
print(experiments)

targets = np.zeros((6,2))

for exp in experiments:
    if len(exp['targets']) > 0:
        for target_idx, target in enumerate(exp['targets']['insertion_targets']['reticle_space']):
            targets[target_idx,0] = target['x']
            targets[target_idx,1] = target['y']
print(targets)
assert(targets.shape[0] == 6)
assert(targets.shape[1] == 2)
```

**Correct target coordinates for shift in agar**

```python
bottom_surface_height = 1.2 #distance from bottom of plastic coverslip to brain surface, MAKE SURE THIS MATCHES NUMBER USED FOR WINDOW CREATION (default is 2.2, subtract 0.5 for each additional spacer coverslip)
top_surface_height = bottom_surface_height + 0.50

calibration_distance_through_agar = 2.2/np.cos(15*np.pi/180) #distance probe traveled through agar during calibration
distance_through_agar = bottom_surface_height/np.cos(15*np.pi/180) #distance probe will travel through agar for this window
distance_ratio = distance_through_agar/calibration_distance_through_agar #correction factor for agar offsets
#print(distance_ratio)
if os.path.isfile('agar_offsets.npy'):
    print('using agar-corrected coords')
    agar_offsets = 1000*distance_ratio*np.load('agar_offsets.npy')
    agar_offsets[:, 0] = -agar_offsets[:, 0] #invert x coordinate to align with josh's axes
    
    print('agar offsets')
    print(agar_offsets)
    
    shifted_targets = np.copy(targets)
    shifted_targets = shifted_targets + agar_offsets #correct target coords for agar drift
    targets_to_use = shifted_targets
else:
    targets_to_use = targets

print(targets_to_use)
```

```python
target_coords = {}
areas = ('AM', 'PM', 'V1', 'LM', 'AL', 'RL')

for idx, area in enumerate(areas):
    target_coords[area] = [np.round(targets_to_use[idx][0]), np.round(targets_to_use[idx][1]), 0]

bottom_surface_height = 2.2
top_surface_height = bottom_surface_height + 0.50

fig = plt.figure(figsize=(7,7))

angles = np.linspace(0,2*np.pi,30)
r = 5
xp = r*np.cos(angles)
yp = r*np.sin(angles)

fname_out = mouse_num + '.eps'

target_structures = ('AM', 'PM', 'V1', 'LM', 'AL', 'RL')
probes = ('A', 'B', 'C', 'D', 'E', 'F')
colors = ('red', 'orange', 'brown', 'green', 'blue', 'purple')

#print(" ")
print(mouse_num)
#print(" ")
print("reticle coordinate insertion points")
for key in target_coords:
    print(key + ": " + str(-target_coords[key][1])+",      " + str(target_coords[key][0])+",      "+ str(target_coords[key][2]))
print(" ")
print("probe coordinate insertion points")
for probe_idx, probe in enumerate(probes):

    origin, R = fit_params(global_coords[probe_idx,:,:], measured_coords[probe_idx,:,:])

    target = target_coords[target_structures[probe_idx]]
    target = np.array(target).T/1000.
    target[0] = -target[0] # invert xcoord
    
    coords = (np.dot(target + origin, R) + np.array([3,3,3]))
    
    print('probe ' + probe + ' (' + target_structures[probe_idx] + '): ' + str(int(coords[0]*1000)) + ', ' + str(int(coords[1]*1000)) + ', ' + str(int(coords[2]*1000)))

    print(extractAngles(R))
    print(-origin*1000)
    print(" ")
    
    invRotMat = np.linalg.inv(R)
    local_pt = coords - np.array([3,3,3])
    insertion_pt = np.dot(local_pt, invRotMat) - origin
    local_pt_bottom = local_pt + [0,0,-bottom_surface_height-target[2]]
    global_pt_bottom = np.dot(local_pt_bottom, invRotMat) - origin

    local_pt_top = local_pt + [0,0,-top_surface_height-target[2]]
    global_pt_top = np.dot(local_pt_top, invRotMat) - origin

    plt.plot(insertion_pt[0],insertion_pt[1],'.',color=colors[probe_idx])
    plt.plot([global_pt_bottom[0],global_pt_top[0]],[global_pt_bottom[1],global_pt_top[1]],colors[probe_idx])
    
for x in range(-5,6):
    for y in range(-5,6):
        plt.plot(x,y,'+',color='0.9')

plt.plot(xp, yp, '--k')
plt.plot([0,0],[-0.5,0.5],'-k')
plt.plot([-0.5,0.5],[0,0],'-k')
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.axis('off')

plt.savefig(fname_out)
```

```python
mouse_num = '419113'

window_list = [
    '412809',
    '415148',
    '412117',
    '415149',
    '406807',
    '405751',
    '412794',
    '404570',
    '412795',
    '406805',
    '412796',
    '408528',
    '409096',
    '403468',
    '404568',
    '397075',
    '397077',
    '404552',
    '398263',
    '404551',
    '412804',
    '410343',
    '412799',
    '412800',
    '412803',
    '408153',
    '410344',
    '408155',
    '407972',
    '407972',
    '412802',
    '408155',
    '408152',
    '403407',
    '412792',
    '404571',
    '404555',
    '404569',
    '398261',
    '404573',
    '419116',
    '419118',
    '419115',
    '419114',
    '419112',
    '411127',
    '410314',
    '410315',
    '419119'
]

def get_target_coords(mouse_num):
    json_string = request.urlopen("http://lims2/specimens/isi_experiment_details/" + mouse_num + ".json").read()
    info = json.loads(json_string)

    experiments = info[0]['isi_experiments']

    targets = np.zeros((6,2))

    for exp in experiments:
        if len(exp['targets']) > 0:
            for target_idx, target in enumerate(exp['targets']):
                targets[target_idx,0] = target['x']
                targets[target_idx,1] = target['y']

    assert(targets.shape[0] == 6)
    assert(targets.shape[1] == 2)

    target_coords = {}
    areas = ('AM', 'PM', 'V1', 'LM', 'AL', 'RL')

    for idx, area in enumerate(areas):
        add_coords = np.array([np.round(targets[idx][0]), np.round(targets[idx][1]), 0])
        try:
            target_coords = np.vstack((target_coords,add_coords))
        except Exception as E:
            target_coords = add_coords
    if mouse_num == '419113':
        target_coords = np.array(    
            [[339,-302,0],
    [905,-40,0],
    [-355,1298,0],
    [-1710,1239,0],
    [-1810,772,0],
    [-1056,-98,0]])
    return target_coords

desired_coords = get_target_coords(mouse_num)
print(mouse_num)
print(desired_coords)
best_match = [10000,10000,10000]
best_match_mouse = False
best_match_coords = False
for mouse in window_list:
    try:
        coords = get_target_coords(mouse)
    except Exception as E:
        print('could not find ',mouse)
    norm = np.linalg.norm(desired_coords - coords)
    try:
        if norm < best_match:
            best_match = norm
            best_match_mouse = mouse
            best_match_coords = coords
    except Exception as E:
        best_match = norm
        best_match_mouse = mouse
        best_match_coords = coords
            
print('Best match mouse: ', best_match_mouse)
print('norm of residuals: ', best_match)
print('coordinares: ')
print(best_match_coords)
```
